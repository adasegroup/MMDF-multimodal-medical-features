```python
import pandas as pd 
import numpy as np

%pylab inline
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt


table = pd.read_csv('/notebook/foundation_of_data_science/multi_modal_data.csv')
```

## Feature correlation to existance\absense the desease

```python
from extract_features import get_feature, get_feature_heart
```

## Age

```python
pos_age, neg_age = get_feature('age', table)
pos_age1, neg_age1 = get_feature_heart('age', table)
```

```python
np.max(pos_age), np.min(pos_age), np.mean(pos_age), np.max(neg_age), np.min(neg_age), np.mean(neg_age)
```

```python
%pylab inline
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt

# the histogram of the data
plt.hist(pos_age, normed=1, facecolor='green', alpha=0.25, label='positive fibrillation: min = 18, max = 94, mean = 61.61')
plt.hist(neg_age, normed=1, facecolor='blue', alpha=0.25, label='negative fibrillation: min = 18, max = 95, mean = 64.71')

plt.xlabel('Age')
plt.ylabel('Number of pos/neg')
plt.title(r'$\mathrm{Histogram\ of\ IQ:}\ \mu=100,\ \sigma=15$')

plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.hist(pos_age1, normed=1, facecolor='green', alpha=0.25, label='positive hf')
plt.hist(neg_age1, normed=1, facecolor='blue', alpha=0.25, label='negative hf')


# add a 'best fit' line
#y = mlab.normpdf( bins, mu, sigma)
#l = plt.plot(bins, y, 'r--', linewidth=1)

plt.xlabel('Age')
plt.ylabel('Number of pos/neg')
plt.title(r'$\mathrm{Histogram\ of\ IQ:}\ \mu=100,\ \sigma=15$')

plt.grid(True)
plt.legend()
plt.show()
```
